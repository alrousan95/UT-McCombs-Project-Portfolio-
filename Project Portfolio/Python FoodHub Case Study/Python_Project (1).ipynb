# Question 12: Multivariate analyis

```python
df.corr()
```

```python
sns.heatmap(df.corr(), annot = True, cmap='YlGnBu');
```

A multivariate analysis is based in observation and analysis of more than one statistical outcome variable at a time. In design and analysis, the technique is used to perform trade studies across multiple dimensions while taking into account the effects of all variables on the responses of interest. As we can observe through the correlation chart and heatmap, most of the variables do not have much correlation amongst each other. The variables have either low or negative correlations amongst each other meaning they are not invertly directed with one another and so the data is random. This could be due to the abundance random outcome of variables in the data and so it is hard to configure a correlation. If we were to look at the heatmap, the data correlation is givin a vizualization. In a heat map, an individual square contains the correlations between two data variables, the darker the square the greater the correlation, the lighter the less. In the heat map, it can be observed there is a lack of correlation between the variables, and in some instances there are negative correlations. In conclusion, the multivariate analysis determines there is little to negative correlation between the data variables. 

# Question 13: The company wants to provide a promotional offer in the advertisement of the restaurants. The condition to get the offer is that the restaurants must have a rating count of more than 50 and the average rating should be greater than 4.

```python
df[df['rating']>4].value_counts(['restaurant_name'])
```

When we first input the code, there is an error. 'Ratings' is not an interger, but instead an object data type and we have to convert the category into a float data type. 

Python is indicating there are values 'not given' but python computes this into  an object and we have to convert the 'not given' value into a nan variable. 

```python
df['rating']=df['rating'].replace(['Not given'], np.nan)
```

This function allows to replace the 'Not given' value to nan. 

```python
df['rating'] = df['rating'].astype(float)
```

Following the function, command python to convery the data type from an object to float, hence creating the data set into an interger. 

```python
df[df['rating']>4].value_counts(['restaurant_name']).head()
```

Once the data set was changed, we are able to conclude Shake Shack and The Meatball Shop will be offered promotional advertisment. 

# Question 14: The company charges the restaurant 25% on the orders having cost greater than 20 dollar and 15% on the orders having cost greater than 5 dollars, while locating the net revenue. 

```python
x = df[df['cost_of_the_order']>20].sum()
revenue = x['cost_of_the_order']
print("The net revenue is", round(revenue * .75, 2), "dollars")
```

```python
y = df[df['cost_of_the_order'].between(5,20)].sum()
lessRevenue = y['cost_of_the_order']
print("The net revenue is", round(lessRevenue * .85, 2), "dollars")
```

The net revenue for orders costing more than 20 dollars is 11,066.18 dollars while the net revenue for orders between 5 dollars and 20 dollars is 14,039.59 dollars.

# Question 15: Finding the percentage of food delivered in 60 minutes. 

```python
df['time'] = df['food_preparation_time'] + df['delivery_time']
x = df[df['time']>60].value_counts().sum()
y = df['time'].value_counts().sum()
print("The percentage of food prepared and delivered averaging more than 60 minutes is", round(x/y, 2),"percent")
```
